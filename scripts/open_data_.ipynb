```python
import csv
```

```python
with open("translations.csv", "r") as file:
    data = file.read()

# Remplacement des caractères " ;" par ","
data_modified = data.replace(";", ",")

# Sauvegarde des données modifiées dans le même fichier CSV
with open("translations.csv", "w") as file:
    file.write(data_modified)
```
